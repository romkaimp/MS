```python
import numpy as np
from scipy.linalg import expm
```

```python
alpha = 1/3
Q = np.array([[-3,  1, 0, 0, 1, 0 ,1 ,0, 0],
              [2,  -7, 0, 5, 0, 0 ,0 ,0, 0],
              [0,  0, -3, 0, 0, 0 ,0 ,3, 0],
              [0,  0, 0, -10, 0, 4 ,6 ,0, 0],
              [0,  0, 0, 0, -3, 3 ,0 ,0, 0],
              [0,  0, 2, 0, 2, -4 ,0 ,0, 0],
              [0,  0, 0, 0, 0, 0 ,-4 ,0, 4],
              [0,  0, 4, 4, 0, 0 ,0 ,-8, 0],
              [0,  0, 0, 0, 0, 0 ,5 ,3, -8],])
P0 = np.array([alpha, 1 - alpha, 0, 0, 0, 0 ,0, 0, 0])  # Начальное состояние
t = 100  # Время эволюции

P_t = P0 @ expm(Q * t)
print(P_t)
```

```python
from scipy.integrate import solve_ivp

def kolmogorov_forward(t, P, Q):
    return P @ Q

Q = np.array([[-3,  1, 0, 0, 1, 0 ,1 ,0, 0],
              [2,  -7, 0, 5, 0, 0 ,0 ,0, 0],
              [0,  0, -3, 0, 0, 0 ,0 ,3, 0],
              [0,  0, 0, -10, 0, 4 ,6 ,0, 0],
              [0,  0, 0, 0, -3, 3 ,0 ,0, 0],
              [0,  0, 2, 0, 2, -4 ,0 ,0, 0],
              [0,  0, 0, 0, 0, 0 ,-4 ,0, 4],
              [0,  0, 4, 4, 0, 0 ,0 ,-8, 0],
              [0,  0, 0, 0, 0, 0 ,5 ,3, -8],])
P0 = np.array([alpha, 1 - alpha, 0, 0, 0, 0 ,0, 0, 0])
t_span = (0, 5)
t_eval = np.linspace(0, 5, 100)

solution = solve_ivp(kolmogorov_forward, t_span, P0, args=(Q,), t_eval=t_eval)

import matplotlib.pyplot as plt
plt.plot(solution.t, solution.y[0], label="P1(t)")
plt.plot(solution.t, solution.y[1], label="P2(t)")
plt.plot(solution.t, solution.y[2], label="P3(t)")
plt.plot(solution.t, solution.y[3], label="P4(t)")
plt.plot(solution.t, solution.y[4], label="P5(t)")
plt.plot(solution.t, solution.y[5], label="P6(t)")
plt.plot(solution.t, solution.y[6], label="P7(t)")
plt.plot(solution.t, solution.y[7], label="P8(t)")
plt.plot(solution.t, solution.y[8], label="P9(t)")
plt.legend()
plt.show()
```
