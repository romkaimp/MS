```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
t = np.linspace(0, 1, n)
X = np.random.choice([-1, 1], size=n, p=[0.5, 0.5])
W = np.cumsum(X) / np.sqrt(n)
```

```python
plt.plot(t, W, label="Случайное блуждание")
plt.xlabel("Время, с")
plt.ylabel("W_n(t)")
plt.title("Сгенерированный винеровский процесс")
plt.grid()
plt.show()
```
