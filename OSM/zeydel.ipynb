```python
import numpy as np
from matplotlib import pyplot as plt
n = 53
N = 4
alpha = 0.003 * (n-50)
```

```python
def y(x):
    return (N + 5.2 + (-1)**N*alpha)*x**3 - (2*N**2+10.4*N+(-1)**(N+1)*alpha)*x**2-N**2*(N+5.2)*(x-2*N)+(-1)**N*alpha
def y_diff(x):
    return 3* (N + 5.2 + (-1)**N*alpha)*x**2 - 2*(2*N**2+10.4*N+(-1)**(N+1)*alpha)*x - N**2*(N+5.2)
```

```python
x_range = np.linspace(-5, 10, 100)
z = [y(x) for x in x_range]
plt.grid(True)
plt.plot(x_range, z, 'b')
plt.show()
```

```python
x1, x2, x3, x4 = -5, 0, 6, 10
```

```python
print("Метод касательных для левого корня:")
xk1 = -1
xk2 = -2
print("x1 =", xk1, "x2 =", xk2)
iters = 0
while abs(y(xk2)) > 0.0001:
    iters += 1
    print("итерация:", iters)
    m = xk2 
    xk2 = xk1 - 1/y_diff(xk1)*y(xk1)
    xk1 = m
    print("x1 =", f"{xk1:.2f}", "x2 =", f"{xk2:.2f}")
    print("ошибка:", f"{abs(y(xk2)):.4f}")
print("x:", f"{xk2:.2f}", "итераций:", iters)
print("Проверка: y(x)=", f"{y(xk2):.2f}")
plt.grid(True)
plt.plot(x_range, z, 'b')
plt.scatter([xk2, ], [0, ], color='r')
plt.show()
```

```python
#  метод деления отрезка пополам 
a, b = x2, x3
print("a=", a, "b=", b)
iters = 0
while abs(y((a + b)/2)) > 0.0001:
    iters += 1
    print("итерация:", iters)
    if y( (b+a)/2 ) > 0:
        a = (a+b)/2
    else:
        b = (a+b)/2
    print("a=", f"{a:.3f}", "b=", f"{b:.3f}")
    print("ошибка:", f'{abs(y((a + b)/2)):.4f}')
print("x:", f"{a:.3f}", "итераций:", iters)
print("Проверка: y(x)=", f'{y((a + b)/2):.3f}')
plt.grid(True)
plt.plot(x_range, z, 'b')
plt.scatter([a, ], [0, ], color='r')
plt.show()
```

```python
#  Метод секущих
xk1 = 7
xk2 = 10
print("x1=", xk1, "x2 =", xk2)
iters = 0
while abs(y(xk2)) > 0.0001:
    iters += 1
    print("итерация:", iters)
    m = xk2 
    xk2 = xk2 - (xk2 - xk1)*y(xk2)/(y(xk2)-y(xk1))
    xk1 = m
    print("x1=", f"{xk1:.2f}", "x2 =", f"{xk2:.2f}")
    print("ошибка:", f"{abs(y(xk2)):.4f}")
print("x=", xk2, "итераций:", iters)
print("Проверка: y(x)=", f"{y(xk2):.4f}")
plt.grid(True)
plt.plot(x_range, z, 'b')
plt.scatter([xk2, ], [0, ], color='r')
plt.show()
```
